```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = plt.imread('Virat1.jpg')

plt.imshow(image)
plt.axis()
plt.show()
```

### Load the Haar Cascade Classifier

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascade
```

### Detect faces

```python
faces = face_cascade.detectMultiScale(image)
faces
```

### OpenCV's cv2.rectangle() function expects a writable image array, as it modifies the image in place by drawing the rectangle.

```python
writable_image = image.copy()
```

### Draw rectangles around the detected faces and display the image

```python
for (x, y, w, h) in faces:
    cv2.rectangle(writable_image, (x, y), (x + w, y + h), (255, 0, 0), 2)
    
# Display the output
plt.imshow(writable_image, cmap='gray')
plt.show()
```

## Combined code

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = plt.imread('Virat1.jpg')

# Display the image
print("Before face detection:")

plt.imshow(image)
plt.axis('Off')
plt.show()


# Load the Harcascade Classifier
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Detect the face
faces = face_cascade.detectMultiScale(image)

# Provide the writable image for drawing bounding box
writable_image = image.copy()

# Draw bounding box
for (x, y, w, h) in faces:
    cv2.rectangle(writable_image, (x, y), (x+w, y+h), (255,0,0), 2)

# Display the output
print('After face Detection:')
plt.imshow(writable_image)
plt.axis('Off')
plt.show()
```

## Face and Eye detection

```python
import cv2
import matplotlib.pyplot as plt


# Loading face-cascade and eye-cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Reading the image
image = plt.imread('Virat1.jpg')  # Only readable (RGB format)

# Converting the image into writable format
writable_image = image.copy()

# RGB to Grayscale format for face and eye detection
gray = cv2.cvtColor(writable_image, cv2.COLOR_RGB2GRAY)

# Detect Face
faces = face_cascade.detectMultiScale(gray)

for (x, y, w, h) in faces:
    # Drawing bounding box for face
    cv2.rectangle(writable_image, (x,y), (x+w, y+h), (255,0,0), 2)
    
    # ROI for eye detection
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = writable_image[y:y+h, x:x+w]
    
    # Detecting eyes
    eyes = eye_cascade.detectMultiScale(roi_gray)
    
    for (ex, ey, ew, eh) in eyes:
        #Drawing bounding box for eyes
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0,255,0), 2)
    
    
    
plt.imshow(writable_image)
plt.show()
```
